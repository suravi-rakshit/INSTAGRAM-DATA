```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("EXPLORATORY DATA ANALYSIS.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("Impressions")["Profile Visits"].sum()
print(grouped_by)
plt.plot(grouped_by.index,grouped_by.values)
plt.xlabel("Impressions",fontsize=20)
plt.ylabel("Profile Visits",fontsize=20)
plt.title("INSTRAGRAM DATA",fontsize=20)
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("EXPLORATORY DATA ANALYSIS.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("Follows")["Profile Visits"].sum()
print(grouped_by)
plt.pie(grouped_by.values,labels=grouped_by.index)
plt.xlabel("Follows",fontsize=20)
plt.ylabel("Profile Visits",fontsize=20)
plt.title("INSTRAGRAM DATA",fontsize=20)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("EXPLORATORY DATA ANALYSIS.xlsx")
df=pd.DataFrame(data)
print(df.head())
plt.hist(df["Impressions"],bins=10)
plt.xlabel("Impressions",fontsize=20)
plt.ylabel("count",fontsize=20)

plt.title("INSTRAGRAM DATA",fontsize=20)
plt.show()

```
